**Q3.Find out the unique genres of movies in this dataset.**

```python
not_null_genre = df['genres'].dropna()
genres =[]
for i in not_null_genre:
  for j in eval(i):
    genres.append(j['name'])
print(set(genres))
```

**Q5.Which actor has acted in the most number of movies?**

```python
df =df['cast'].dropna()#DEMO
```

```python
actor = df
actor_name  = []
for i in actor[:]:
  for j in eval(i):
    actor_name.append(j['name'])
actor_dataframe = pd.DataFrame(actor_name,columns=['Name'])
actor_count  = actor_dataframe['Name'].value_counts()
actor_count.head()
```
