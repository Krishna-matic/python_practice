```python
arr = np.arange(1,13).reshape(3,4)
print(f'Array shape: {arr.shape}')
print(f'Array Dimension: {arr.ndim}')
print(f'Array size: {arr.size}')
print(f'Array dtype: {arr.dtype}')
arr= arr.astype(np.float32)
print(arr)
arr = arr.T
print(arr)
arr = arr.flatten()
print(arr)
print(f"Final Array: {arr}")
```
